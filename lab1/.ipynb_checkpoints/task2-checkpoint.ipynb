# Задача 2.1.

---
Пусть задано два вектора ответов: y - истинный вектор ответов для некоторой выборки, а также есть вектор ответов $\hat{y}$ - некоторой предсказательной модели. Наблюдатель хочет проверить гипотезу о том, что ровно в 25% случаев модель дает заниженные оценки. Предложите метод проверки данной гипотезы: запишите задачу формально, предложите статистику для решения данной задачи на уровне значимости alpha = 0.05. Также найдите зависимость мощности данного критерия в зависимости от истинного процента заниженных ответов.

Все выкладки должны быть сделаны аналитически, без использования компьютера. (допускается использование компютера для подстановвки численых значений в финальную формулу)

---

Для проверки гипотезы о том, что в 25% случаев модель дает заниженные оценки, будем использовать биномиальный тест.

1. Введем биномиальную случайную величину z - количество заниженных оценок на выборке. Если записать формально, то:
$$z = \sum_{i=1}^n{[y_i > \hat{y_i}]}.$$


2. Далее сформулируем гипотезы относительно распределения случайной величины z: нулевую гипотезу $H_0$: "Доля заниженных оценок равна 0.25 (вероятность $p=0.25$)", и альтернативную гипотезу $H_a$: "Доля заниженных оценок больше или меньше 0.25 (вероятность $p!=0.25$)".



3. Определим уровень значимости alpha из условия задачи: alpha=0.05.


4. Теперь применим биномиальный тест к данной формальной задаче с помощью, например, класса scipy.stats.binom_test:

```
test = scipy.stats.binom_test(z, n, p=0.25, alternative='two-sided')
test.pvalue
```

Здесь z - число заниженных оценок на выборке, n - размер выборки, p = 0.25 - заданная нулевая гипотеза, alternative = "two_sided" - двусторонняя альтернатива (так как мы проверяем неравенство).



5. Интерпретировать результаты теста: если p-value меньше заданного уровня значимости alpha, то отвергаем нулевую гипотезу в пользу двусторонней альтернативной. Если p-value больше alpha, то нет оснований для отвержения нулевой гипотезы.

6. Зависимость мощности данного критерия от истинного процента заниженных ответов найдем тем же эмририческим способом, которым я искал зависимость мощности критерия от коэффициентов alpha и delta в 1 задаче пункты 5 и 7 (task1.ipynb разделы 5 и 7).
